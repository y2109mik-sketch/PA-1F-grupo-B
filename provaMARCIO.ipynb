```python
#Classificação de idade

#indentificar se o usuario é
# <17 menor de idade
# <59 adulto
# <60 idoso

# entrada

idade = int(input("digite sua idade"))


#processamento/saida

if idade <= 17:
  print(f"voce tem {idade} ano de idade, voce é menor de idade")

elif idade <=59:
    print(f" voce tem {idade} ano de idade, voce é um adulto")

    if idade >60:
      print(f"se voce tem {idade} ano de idade, voce é um idoso")

```

```python
# programa de desconto

# < 100 sem desconto
# >=101 e <=500 10% de desconto
# > 500 20% de desconto


# Entrada

compra_original = float(input("digite o valor da sua compra: "))

# Processamento
desconto_percentual = 0.0
valor_desconto = 0.0
compra_final = compra_original

if compra_original >= 101 and compra_original <= 500:
  desconto_percentual = 10.0
  valor_desconto = compra_original * (desconto_percentual / 100)
  compra_final = compra_original - valor_desconto
elif compra_original > 500:
  desconto_percentual = 20.0
  valor_desconto = compra_original * (desconto_percentual / 100)
  compra_final = compra_original - valor_desconto
# Se compra_original for < 101, desconto_percentual permanece 0.0 e valor_desconto 0.0

# Saida

print(f"Valor original da compra: R${compra_original:.2f}")
if desconto_percentual > 0:
  print(f"Desconto aplicado: {desconto_percentual:.0f}% (R${valor_desconto:.2f})")
  print(f"Valor final com desconto: R${compra_final:.2f}")
elif compra_original < 101:
  print(f"Você não recebeu desconto. Valor final: R${compra_final:.2f}")
```
