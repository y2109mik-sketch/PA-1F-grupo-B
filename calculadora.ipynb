```python
# operação aritimetica
# fazer um programa que calcula operações com 02 numeros

# Entrada

a = int(input("digite um numero: "))
b = int(input("digite um numero: "))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b

# saída

print("")
print("-----------------------OPERADORES ARITIMETICOS----------------------")
print("")
print("--------------------------------SOMA--------------------------------")
print("")
print (f"o resultado da SOMA de {a} MAIS {b} é igual é: {soma} ")
print("")
print("")
print("-----------------------OPERADORES ARITIMETICOS----------------------")
print("")
print("------------------------------SUBTRAÇÃO-----------------------------")
print("")
print (f"o resultado da SUBTRAÇÃO de {a} MENOS {b} é igual é: {subtracao} ")
print("")
print("-----------------------OPERADORES ARITIMETICOS----------------------")
print("")
print("----------------------------MULTIPLICAÇÃO---------------------------")
print("")
print (f"o resultado da MULTIPLICAÇÃO de {a} MULTIPLICADO por {b} é igual é: {multiplicacao} ")
print("")
print("-----------------------OPERADORES ARITIMETICOS----------------------")
print("")
print("------------------------------DIVISÃO-------------------------------")
print("")
print (f"o resultado da DIVISÃO de {a} DIVIDIDO por {b} é igual é: {divisao} ")
print("")
print("---------------------------DIVISÃO INTEIRA--------------------------")
print("")
print (f"o resultado da DIVISÃO INTEIRA de {a} DIVIDIDO por {b} é igual é: {div_int} ")
print("")
print("---------------------------RESTO DA DIVISÃO-------------------------")
print("")
print (f"o resto da DIVISÃO de {a} DIVIDIDO por {b} é igual é: {resto} ")
print("")
print("-----------------------------POTENCIACÃO----------------------------")
print("")
print (f"o resultado da POTENCIA de {a} ELEVADO {b} é igual é: {potencia} ")
print("")





```
